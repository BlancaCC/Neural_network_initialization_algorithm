## Neural network

### Neural network definition 
Neural networks are defined by their weight $A, B$ matrix and an activation function. 

For example: 


```python
import numpy as np
from one_layer_neural_network import OneLayerNeuralNetwork
from activation_functions import relu
A = np.array([[1,0,1],[0,1,1]])
B = np.array([[1,0]])
h = OneLayerNeuralNetwork(A, B, relu)
```

### Neural network evaluation 

```python
X = np.array([[1,2], [0,0],[-1,4]])
h.predict(X)
```

# Initialization algorithm 

```python
from learnt_initialization import learnt_initialization
ideal_function = lambda x : x*x*x - x*x +x 
W = 100
X = np.linspace(-5,5,W)
y = np.array(list(map(lambda x: [x], ideal_function(X))))
X = np.array(list(map(lambda x: [x], X)))
for activation_function in ['tanh', 'ramp']:
    for n in range(2,80,20):
        h = learnt_initialization(X, y, n, activation_function)

        import matplotlib.pyplot as plt
        plt.plot(X, h.predict(X))
        plt.plot(X, ideal_function(X))
        plt.title(f'number of neurons = {n}, activation function  = {activation_function}')
        plt.show()
```
